## Bag of Words

```python
from collections import Counter
```

```python
def tokens_mais_frequentes(tokens, tam=10):
    """gera lista dos `tam` tokens mais frequentes com a respectiva contagem"""
    
    return Counter(tokens).most_common(tam)
```

## TD-IDF

Term Frequency - Inverse Document Frequency

```python
from gensim.models.tfidfmodel import TfidfModel
```

```python
def tfdif_palavras_mais_representativas(dicionario, tfidf_doc, tam=10):
    # lista ordenada pelo peso tfidf do termo
    tfidf_desc = sorted(tfidf_doc, key=lambda termo: termo[1], reverse=True)
    
    # lista no tamanho especificado
    tfidf_desc_tam = tfidf_desc[:tam]
    
    # conversão dos tokenids para palavras
    palavras = [(dicionario.get(tokenid), peso) for tokenid, peso in tfidf_desc_tam]
    
    return palavras
```

```python
tfidf = TfidfModel(corpus)
```

```python
for doc in corpus:
    if len(doc) > 0:
        tfidf_doc = tfidf[doc]
        print(tfdif_palavras_mais_representativas(dicionario, tfidf_doc, tam=6))
        print()
```

### NER

```python
#import nltk
#nltk.download('maxent_ne_chunker')
#nltk.download('averaged_perceptron_tagger')
#nltk.download('words')
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
def divide_sentencas(doc_bruto):
    return sent_tokenize(doc_bruto)
```

```python
def tokeniza_sentencas(sentencas):
    return [word_tokenize(sentenca) for sentenca in sentencas]
```

```python
def gera_pos_tag(sentencas_tokenizadas):
    return [nltk.pos_tag(sentenca) for sentenca in sentencas_tokenizadas]
```

```python
def gera_chunks_ner(sentencas_com_pos):
    return list(nltk.ne_chunk_sents(sentencas_com_pos, binary=True))
```

```python
def filtra_chunks_ner(sents_chunks_ner, filtro):
    res = []
    for sentenca in sents_chunks_ner:
        for chunk in sentenca:
            if hasattr(chunk, 'label') and chunk.label() == filtro:
                res.append(chunk)
    return res
```

```python
def imprime_chunks_ner_filtrados(sents_chunks_ner, filtro):
    for el in filtra_chunks_ner(sents_chunks_ner, filtro=filtro):
        print(el)
```

```python
sents = divide_sentencas(texto_bruto)
sents_tks = tokeniza_sentencas(sents)
sents_pos = gera_pos_tag(sents_tks)
sents_chunks_ner = gera_chunks_ner(sents_pos)
```

```python
imprime_chunks_ner_filtrados(sents_chunks_ner, filtro='NE')
```
